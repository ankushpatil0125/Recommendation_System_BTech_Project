```python
df=df.drop(columns=['Crawl Timestamp','Job Salary','Role Category','Functional Area','Industry','Role'])
```

```python
df['Job Experience Required']=df['Job Experience Required'].fillna(0)
df['Experience'] = df['Job Experience Required'].str[::-1].str.extract('(\d+)' ,expand= True)
df['Experience']=df['Experience'].fillna(0)
df['Experience']=df['Experience'].astype(int)
df=df.drop(columns=['Job Experience Required'])
```

```python
df['Job Title']=df['Job Title'].str.strip()
df['Job Title']=df['Job Title'].str.lower()
df['Key Skills']=df['Key Skills'].str.strip()
df['Key Skills']=df['Key Skills'].str.lower()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv=TfidfVectorizer(min_df=3, max_features=None,
                   strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}\S',
                   ngram_range=(1,3),
                   stop_words='english')
df['Job Title']=df['Job Title'].fillna('')
df['Key Skills']=df['Key Skills'].fillna('')
df['Job']=df['Job Title']+df['Key Skills']
```
